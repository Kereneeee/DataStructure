```python
#暖身
#Leedcode 題目
from typing import List
def two_sum(nums:List[int],target:int) -> List[int]:
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if nums[i]+nums[j] == target:
                return [i,j]
print(two_sum([2,7,6,5],9)) 
```

```python
#串列元素相加
def sum(arr,n):
    total = 0               #1
    for i in range(n):      #n
        total += arr[i]     #n
    return total            #1

# 2n + 2
# Big-O = O(n)
```

```python
#矩陣相加
def add(a,b,c,n):
    for i in range(n):                      #n
        for j in range(n):                  #n(n)
            c[i][j] = a[i][j] + b[i][j]     #n(n)

# 2n^2 + n
# Big-O = O(n^2)
```

```python
# 矩陣相乘
def multiply(a,b,c,n):
    for i in range(n):                      #n
        for j in range(n):                  #n(n)
            sum = 0                         #n(n)
            for k in range(n):              #n(n)(n)
                sum = a[i][k] + b[k][j]     #n(n)(n)
            c[i][j] = sum                   #n(n)

# 2n^3 + 3N^2 + n
# Big-O = O(n^3)
```

```python
#循序搜尋
def search(data,target,n):
    for i in range(n):              #n
        if target == data[i]:       #n
            return i                #1

# 2n + 1
# Big-O = O(n)
```
